```python
import os, random,shutil
```

```python
#static class used to copy in a new folder our picture dataset (TRAINING)
```

```python
class Klass(object):

#generator to yield numbers from starting value to final value
    @classmethod
    def my_counter(cls,low, high):
        current = low
        while current < high:
            yield current
            current += 1
#method to make a directory         
    @classmethod
    def mk_dir(cls,a_dir):
        try:
            os.makedirs(a_dir)
        except:
            print("the directory already exist")
            pass
#method to remove the directory  
    @classmethod
    def rmv_dir(cls,a_dir):    
        try:
            shutil.rmtree(a_dir)
        except:
            print("the directory doesn't exist")
            pass
#method to pick random files (100k in total)
    @classmethod
    def randomize(cls,my_dir="C:/Users/212613037/Pictures/ML/chest_web_Copie/",my_dest_dir="D:/KASKAD/Pictures/",my_length=100000):
        cls.rmv_dir(my_dest_dir)     
        cls.mk_dir(my_dest_dir)
        
        for c in cls.my_counter(0,my_length):
            l=random.choice(os.listdir("C:/Users/212613037/Pictures/ML/chest_web_Copie/"))
            l=os.path.join(my_dir,l)
            try:
                shutil.copy(l,my_dest_dir)
            except:
                print('error while copying')
                pass
```
